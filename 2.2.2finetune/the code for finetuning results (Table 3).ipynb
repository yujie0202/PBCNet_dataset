```python
from sklearn.metrics import mean_squared_error
import numpy as np
import pandas as pd
import os
from scipy.stats import spearmanr,pearsonr
import itertools
```

```python
code_path = os.path.abspath('.')
code_path
```

```python
fep1_files = ['PTP1B', 'Jnk1', 'CDK2', 'Thrombin', 'Tyk2', 'p38', 'Bace', 'MCL1']
fep2_files = ['cdk8', 'tnks2', 'hif2a', 'cmet', 'eg5', 'pfkfb3', 'shp2', 'syk']
```

```python
def read_mol2_file(filename):
    atoms = []
    with open(filename, 'r') as f:
        all_line = f.read().split("\n")
        for line_o in all_line:
            line = line_o.rstrip().split()
            if len(line) == 2:
                if line[1] == '9':
                    
                    ind = all_line.index(line_o)
                    
                    line_without_ligands = all_line[ind+1].rstrip().split()
                    line_with_ligands = all_line[ind+2].rstrip().split()

                    return float(line_without_ligands[3]),float(line_without_ligands[5]),float(line_without_ligands[7]), \
                           float(line_with_ligands[3]),float(line_with_ligands[5]),float(line_with_ligands[7])
```

# FEP2

```python
ref_num=[2,3,4,5,6,7,8,9,10]
for ref in ref_num:
    print(ref)

    R = []
    S = []
    P=[]
    S_max = []
    S_min = []
    P_max = []
    P_min = []

    R2 = []
    S2 = []
    P2 = []
    S_max2 = []
    S_min2 = []
    P_max2 = []
    P_min2 = []
    for file in fep2_files:

        chongfu = [i for i in range(0,10)]

        r_ = 0
        r_var = 0
        s_ = 0
        s_var = 0
        p_ = 0
        p_var = 0

        r_2 = 0
        r_var2 = 0
        s_2 = 0
        s_var2 = 0
        p_2 = 0
        p_var2 = 0
        r = []
        s = []
        p = []
        r2 = []
        s2 = []
        p2 = []
        for cf in chongfu:
            rmse,spearman,pearson,rmse2,spearman2,pearson2 =read_mol2_file(
            f"{code_path}/finetune_results/{file}/ref{ref}_results/{cf}_record.txt") 
            r.append(rmse)
            s.append(spearman)
            p.append(pearson)
            r2.append(rmse2)
            s2.append(spearman2)
            p2.append(pearson2)

        r = np.array(r)
        s = np.array(s)
        p = np.array(p)
        r2 = np.array(r2)
        s2 = np.array(s2)
        p2 = np.array(p2)


        r_ += np.mean(r)
        r_var+=np.var(r)
        s_+= np.mean(s)
        s_var+=np.var(s)
        s_min = np.min(s)
        s_max = np.max(s)
        p_+= np.mean(p)
        p_var+=np.var(p)
        p_min = np.min(p)
        p_max = np.max(p)

        r_2 += np.mean(r2)
        r_var2+=np.var(r2)
        s_2+= np.mean(s2)
        s_var2+=np.var(s2)
        s_min2 = np.min(s2)
        s_max2 = np.max(s2)
        p_2+= np.mean(p2)
        p_var2+=np.var(p2)
        p_min2 = np.min(p2)
        p_max2 = np.max(p2)


        R.append(r_)
        S.append(s_)
        P.append(p_)
        S_max.append(s_max)
        P_max.append(p_max)
        S_min.append(s_min)
        P_min.append(p_min)

        R2.append(r_2)
        S2.append(s_2)
        P2.append(p_2)
        S_max2.append(s_max2)
        P_max2.append(p_max2)
        S_min2.append(s_min2)
        P_min2.append(p_min2)

        print(f"{file} without rmse_G {r_} {r_var} spearman {s_} {s_var} pearson {p_} {p_var}")
        print(f"{file}  with   rmse_G {r_2} {r_var2} spearman {s_2} {s_var2} pearson {p_2} {p_var2}")
    print(f"fep2 rmseG {np.mean(R)} spearman {np.mean(S)} pearson {np.mean(P)}")
    print(f"fep2 rmseG {np.var(R)} spearman {np.var(S)} pearson {np.var(P)}")
    
    print(f"fep2 min spearman {np.mean(S_min)} pearson {np.mean(P_min)}")
    print(f"fep2 max spearman {np.mean(S_max)} pearson {np.mean(P_max)}")
    print(f"fep2 rmseG {np.mean(R2)} spearman {np.mean(S2)} pearson {np.mean(P2)}")
    print(f"fep2 min spearman {np.mean(S_min2)} pearson {np.mean(P_min2)}")
    print(f"fep2 max spearman {np.mean(S_max2)} pearson {np.mean(P_max2)}")
```
