```python
import os
import pandas as pd
import numpy as np
```

```python
systems = os.listdir('/home/yujie/leadopt/robust/')
systems
```

```python
for i in [0,1,2,3,4,5,6,7]:
    names = os.listdir(f'/home/yujie/leadopt/selet/{systems[i]}/')
    names

    s = []
    p = []

    for name in names:
        re = pd.read_csv(f"/home/yujie/leadopt/robust/{systems[i]}/{name}/prefinetune_results.csv")
        print(systems[i])
        print(name)
        print(f"s: {np.mean(re.spearman)}")
        print(f"p: {np.mean(re.pearson)}")
        s.append(np.mean(re.spearman))
        p.append(np.mean(re.pearson))
    print(np.mean(s))
    print(np.mean(p))
    print(np.var(s))
    print(np.var(p))
    print("")
    print("")
```
